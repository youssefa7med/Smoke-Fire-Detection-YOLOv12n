```python
import cv2
from ultralytics import YOLO
import pygame
import requests
import time
from datetime import datetime

token = "8391703782:AAE0Di6EbKc9pkrpHr8LvZ-hW8Y6n1JCXOk"
chat_id = "877368153"

def send_telegram_alert(frame, class_name):
    url = f"https://api.telegram.org/bot{token}/sendPhoto"

    _, buffer = cv2.imencode('.jpg', frame)
    files = {'photo': buffer.tobytes()}

    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    caption_text = (
        f"ðŸš¨ *FIRE ALERT!* ðŸš¨\n"
        f"ðŸ”¥ Detected: {class_name}\n"
        f"ðŸ“ Location: Lab / Camera 1\n"
        f"â±ï¸ Detection Time: {now}\n"
        f"âš ï¸ Please take immediate action!"
    )

    requests.post(
        url,
        data={"chat_id": chat_id, "caption": caption_text, "parse_mode": "Markdown"},
        files=files
    )
    print("ðŸ”¥ Fire Alert sent to Telegram!")

cooldown_seconds = 30
last_sent_time = 0

model = YOLO("D:\\data science\\note.books\\Smoke-Fire-Detection-YOLOv12n\\best.pt")

pygame.mixer.init()
alarm_sound = "D:\\data science\\note.books\\Smoke-Fire-Detection-YOLOv12n\\minion_fire_alarm.mp3"
pygame.mixer.music.load(alarm_sound)
alarm_playing = False

video_file = "D:\\data science\\note.books\\Smoke-Fire-Detection-YOLOv12n\\Test Video.mp4"
cap = cv2.VideoCapture(video_file)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    annotated_frame = results[0].plot()

    detected_classes = results[0].names
    labels = results[0].boxes.cls if results[0].boxes is not None else []

    fire_detected = False
    detected_name = ""

    for cls in labels:
        class_name = detected_classes[int(cls)]
        if class_name.lower() in ["fire", "smoke"]:
            fire_detected = True
            detected_name = class_name
            break

    current_time = time.time()

    if fire_detected:
        if not alarm_playing:
            pygame.mixer.music.play(-1)
            alarm_playing = True

        if current_time - last_sent_time >= cooldown_seconds:
            send_telegram_alert(frame, detected_name)
            last_sent_time = current_time

    elif not fire_detected and alarm_playing:
        pygame.mixer.music.stop()
        alarm_playing = False

    cv2.imshow("YOLOv12n Fire Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
pygame.mixer.music.stop()
```
